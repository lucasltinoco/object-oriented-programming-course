### 8) Escreva uma classe Coordenada com atributos x e y, e métodos para mostrar as coordenadas, calcular a distancia para outra coordenada, comparar coordenadas, mostrar no formato coordenada polar.


```python
import math
import numpy


class Coordenada():

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def mostra_coordenada(self):
        return [self.x, self.y]

    def calcula_distancia(self, coordenada_vizinha):
        return math.sqrt(math.pow(coordenada_vizinha.x - self.x, 2) + math.pow(coordenada_vizinha.y - self.x, 2))

    def compara(self, coordenada_vizinha):
        return self.x == coordenada_vizinha.x and self.y == coordenada_vizinha.y

    def converte_para_polar(self):
        raio = math.sqrt(math.pow(self.x, 2) + math.pow(self.y, 2))
        theta = numpy.arctan2(self.y, self.x)
        angulo_graus = 180 * (theta / math.pi)

        return {raio, theta, angulo_graus}


def main():
    ponto1 = Coordenada(0, 1)
    ponto2 = Coordenada(0, 10)

    print(ponto1.mostra_coordenada())
    print(ponto1.calcula_distancia(ponto2))
    print(ponto1.compara(ponto2))
    print(ponto1.converte_para_polar())


main()
```

### 9) Escreva classes para as seguintes formas: quadrado, retângulo e círculo


```python
import math


class Quadrado():

    def __init__(self, lado):
        self.lado = lado

    def calcula_perimetro(self):
        return self.lado * 4

    def calcula_area(self):
        return self.lado * self.lado / 2


class Retangulo():

    def __init__(self, lado_maior, lado_menor):
        self.lado_maior = lado_maior
        self.lado_menor = lado_menor

    def calcula_perimetro(self):
        return self.lado_maior * 2 + self.lado_menor * 2

    def calcula_area(self):
        return self.lado_maior * self.lado_menor / 2


class Circulo():

    def __init__(self, raio):
        self.raio = raio

    def calcula_perimetro(self):
        return 2 * math.pi * self.raio

    def calcula_area(self):
        return math.pi * self.raio * self.raio


def main():
    quadrado = Quadrado(2)
    print(quadrado.calcula_perimetro())
    print(quadrado.calcula_area())

    retangulo = Retangulo(2, 3)
    print(retangulo.calcula_perimetro())
    print(retangulo.calcula_area())

    circulo = Circulo(8)
    print(circulo.calcula_perimetro())
    print(circulo.calcula_area())


main()
```

### 10) Escreva uma classe Fracao que armazena dois inteiros, numerador e denominador.

    a. Implemente metodos para somas, subtração, multiplicação e divisão de duas frações
    b. Implemente o método que imprime uma fração no formato numerador denominador
    c. Implemente um método que inverte a fração
    d. Implemente um método que retorna a fração em valor real
    e. Implemente um método que cria uma fração (numerador/denominador) a partir de um número real


```python
class Fracao():
    def __init__(self, numerador, denominador):
        self.numerador = numerador
        self.denominador = denominador

    def __mdc__(self, num, den):
        while num % den != 0:
            num_anterior = num
            den_anterior = den

            num = den_anterior
            den = num_anterior % den_anterior
        return den

    def soma(self, outra_fracao):
        novo_numerador = self.numerador * outra_fracao.denominador + \
            self.denominador * outra_fracao.numerador
        novo_denominador = self.denominador * outra_fracao.denominador

        denominador_comum = self.__mdc__(novo_numerador, novo_denominador)

        return Fracao(novo_numerador // denominador_comum, novo_denominador // denominador_comum)

    def subtrai(self, outra_fracao):
        novo_numerador = self.numerador * outra_fracao.denominador - \
            self.denominador * outra_fracao.numerador
        novo_denominador = self.denominador * outra_fracao.denominador

        denominador_comum = self.__mdc__(novo_numerador, novo_denominador)

        return Fracao(novo_numerador // denominador_comum, novo_denominador // denominador_comum)

    def multiplica(self, outra_fracao):
        novo_numerador = self.numerador * outra_fracao.numerador
        novo_denominador = self.denominador * outra_fracao.denominador

        denominador_comum = self.__mdc__(novo_numerador, novo_denominador)

        return Fracao(novo_numerador // denominador_comum, novo_denominador // denominador_comum)

    def divide(self, outra_fracao):
        novo_numerador = self.numerador * outra_fracao.denominador
        novo_denominador = self.denominador * outra_fracao.numerador

        denominador_comum = self.__mdc__(novo_numerador, novo_denominador)

        return Fracao(novo_numerador // denominador_comum, novo_denominador // denominador_comum)

    def imprime(self):
        print(self.numerador, "/", self.denominador)

    def inverte(self):
        novo_numerador = self.denominador
        novo_denominador = self.numerador

        denominador_comum = self.__mdc__(novo_numerador, novo_denominador)

        return Fracao(novo_numerador // denominador_comum, novo_denominador // denominador_comum)

    def converte_para_real(self):
        return self.numerador / self.denominador

    def converte_para_fracao(self, real):
        if int(real) == real:
            return Fracao(int(real), 1)

        real_str = str(real)
        real_split = real_str.split('.')

        novo_numerador = int(''.join(real_split))
        novo_denominador = 10 ** len(real_split[1])

        denominador_comum = self.__mdc__(novo_numerador, novo_denominador)

        return Fracao(novo_numerador // denominador_comum, novo_denominador // denominador_comum)


def main():
    fracao1 = Fracao(2, 4)
    fracao2 = Fracao(1, 4)
    soma = fracao1.soma(fracao2)
    subtracao = fracao1.subtrai(fracao2)
    multiplicacao = fracao1.multiplica(fracao2)
    divisao = fracao1.divide(fracao2)
    invertida = fracao1.inverte()
    real = fracao1.converte_para_real()
    real_para_fracao = fracao1.converte_para_fracao(34.5)

    print(soma.imprime())
    print(subtracao.imprime())
    print(multiplicacao.imprime())
    print(divisao.imprime())
    print(invertida.imprime())
    print(real)
    print(real_para_fracao.imprime())


main()
```

### Exercício 1: Banco

O banco Tatu, moderno e eficiente, precisa de um novo programa para controlar o
saldo de seus correntistas. Cada conta corrente pode ter um ou mais clientes
como titular. O banco controla apenas o nome e o telefone de cada cliente. A
conta corrente apresenta um saldo e uma lista de operações de saques e
depósitos. Quando o cliente fizer um saque, diminuiremos o saldo da conta
corrente. Quando ele fizer um depósito, aumentaremos o saldo. O banco oferece
também contas especiais, com limite especial além do saldo, e conta poupança,
que oferece um rendimento mensal sempre que o saldo na conta completa um mês.
Evidentemente é necessário oferecer os clientes a possibilidade de verificar
saldos, extratos e um resumo com todas as informações da conta e seus
respectivos clientes.


```python
class Cliente:
    def __init__(self, nome, telefone):
        self.nome = nome
        self.telefone = telefone


class Conta:
    def __init__(self, clientes, numero, saldo=0):
        self.saldo_inicial = saldo
        self.saldo = saldo
        self.clientes = clientes
        self.numero = numero
        self.operacoes = []

    def saque(self, valor):
        if self.saldo >= valor:
            self.saldo -= valor
            self.operacoes.append(["SAQUE", valor])
        return valor

    def deposito(self, valor):
        self.saldo += valor
        self.operacoes.append(["DEPÓSITO", valor])

    def imprime_saldo(self):
        print(f"Saldo: {self.saldo}\n")

    def extrato(self):
        print(f"Extrato da conta {self.numero}\n")
        print(f"Saldo inicial {self.saldo_inicial}\n")
        for operacao in self.operacoes:
            print(f"{operacao[0]} {operacao[1]}\n")
        print(f"Saldo: {self.saldo}\n")

    def resumo(self):
        print(f"Número da conta: {self.numero}\n")
        print(f"Clientes: \n")
        for cliente in self.clientes:
            print(f"{cliente.nome} {cliente.telefone}\n")
        self.imprime_saldo()
        self.extrato()


class Banco:
    def __init__(self, nome):
        self.nome = nome
        self.contas = []

    def adiciona_conta(self, conta):
        self.contas.append(conta)

    def imprime_contas(self):
        for conta in self.contas:
            conta.resumo()


class ContaEspecial(Conta):
    def __init__(self, clientes, numero, saldo=0, limite=0):
        Conta.__init__(self, clientes, numero, saldo)
        self.limite = limite

    def saque(self, valor):
        if self.saldo + self.limite >= valor:
            self.saldo -= valor
            self.operacoes.append(["SAQUE", valor])


class ContaPoupanca(Conta):
    def __init__(self, clientes, numero, saldo=0, rendimento_mensal=0):
        Conta.__init__(self, clientes, numero, saldo)
        self.rendimento_mensal = rendimento_mensal

    def rende(self):
        valor = self.saldo * self.rendimento_mensal
        self.saldo += valor
        self.operacoes.append(["RENDIMENTO MENSAL", self.saldo])


def main():
    fulano = Cliente("Fulano", "(48) 99999-9999")
    beltrano = Cliente("Beltrano", "(21) 99999-9999")
    ciclano = Cliente("Ciclano", "(47) 99999-9999")

    banco_tatu = Banco("Tatu")

    conta_corrente = Conta([fulano, beltrano], 1, 100)
    # conta_corrente.deposito(1000)
    # conta_corrente.saque(200)
    # conta_corrente.resumo()

    banco_tatu.adiciona_conta(conta_corrente)

    conta_especial = ContaEspecial([fulano, beltrano, ciclano], 2, 100, 100)
    # conta_especial.saque(150)
    # conta_especial.resumo()

    banco_tatu.adiciona_conta(conta_especial)

    conta_poupanca = ContaPoupanca([fulano, beltrano, ciclano], 3, 100, 0.1)
    # conta_poupanca.imprime_saldo()
    # conta_poupanca.rende()
    # conta_poupanca.imprime_saldo()

    banco_tatu.adiciona_conta(conta_poupanca)

    banco_tatu.imprime_contas()


main()
```

### Exercício 2 Biblioteca

Crie as classes necessárias para um sistema de gerenciamento de uma biblioteca.
Os bibliotecários deverão preencher o sistema com o título do livro, os autores,
o ano, a editora, a edição e o volume. A biblioteca também terá um sistema de
pesquisa (outro software), portanto será necessário conseguir acessar os
atributos típicos de pesquisa (nome, autor, …).


```python
class Capitulo:
    def __init__(self, numero: int, titulo: str):
        self.__numero = numero
        self.__titulo = titulo

    @property
    def numero(self):
        return self.__numero

    @property
    def titulo(self):
        return self.__titulo

    @numero.setter
    def numero(self, numero):
        self.__numero = numero

    @titulo.setter
    def titulo(self, titulo):
        self.__titulo = titulo


class Editora:
    def __init__(self, codigo: int, nome: str):
        self.__codigo = codigo
        self.__nome = nome

    @property
    def codigo(self):
        return self.__codigo

    @property
    def nome(self):
        return self.__nome

    @codigo.setter
    def codigo(self, codigo):
        self.__codigo = codigo

    @nome.setter
    def nome(self, nome):
        self.__nome = nome


class Autor:
    def __init__(self, codigo: int, nome: str):
        self.__codigo = codigo
        self.__nome = nome

    @property
    def codigo(self):
        return self.__codigo

    @property
    def nome(self):
        return self.__nome

    @codigo.setter
    def codigo(self, codigo):
        self.__codigo = codigo

    @nome.setter
    def nome(self, nome):
        self.__nome = nome


class Livro:
    def __init__(self, codigo: int, titulo: str, ano: int, editora: Editora, autor: Autor, numeroCapitulo: int, tituloCapitulo: str, edicao: int, volume: int):
        self.__codigo = codigo
        self.__titulo = titulo
        self.__ano = ano
        self.__editora = editora
        self.__autor = autor
        self.__numeroCapitulo = numeroCapitulo
        self.__tituloCapitulo = tituloCapitulo
        self.__autores = []
        self.__capitulos = []
        self.__edicao = edicao
        self.__volume = volume

    @property
    def codigo(self):
        return self.__codigo

    @property
    def titulo(self):
        return self.__titulo

    @property
    def ano(self):
        return self.__ano

    @property
    def editora(self):
        return self.__editora

    @property
    def autor(self):
        return self.__autor

    @property
    def numeroCapitulo(self):
        return self.__numeroCapitulo

    @property
    def tituloCapitulo(self):
        return self.__tituloCapitulo

    @property
    def autores(self):
        return self.__autores

    @property
    def capitulos(self):
        return self.__capitulos

    @property
    def edicao(self):
        return self.__edicao

    @property
    def volume(self):
        return self.__volume

    @codigo.setter
    def codigo(self, codigo):
        self.__codigo = codigo

    @titulo.setter
    def titulo(self, titulo):
        self.__titulo = titulo

    @ano.setter
    def ano(self, ano):
        self.__ano = ano

    @editora.setter
    def editora(self, editora):
        self.__editora = editora

    @autor.setter
    def autor(self, autor):
        self.__autor = autor

    @numeroCapitulo.setter
    def numeroCapitulo(self, numeroCapitulo):
        self.__numeroCapitulo = numeroCapitulo

    @tituloCapitulo.setter
    def tituloCapitulo(self, tituloCapitulo):
        self.__tituloCapitulo = tituloCapitulo

    @autores.setter
    def autores(self, autores):
        self.__autores = autores

    @capitulos.setter
    def capitulos(self, capitulos):
        self.__capitulos = capitulos

    @edicao.setter
    def edicao(self, edicao):
        self.__edicao = edicao

    @volume.setter
    def volume(self, volume):
        self.__volume = volume

    def incluirAutor(self, autor: Autor):
        if isinstance(autor, Autor) and autor not in self.__autores:
            self.__autores.append(autor)

    def excluirAutor(self, autor: Autor):
        self.__autores = [
            a for a in self.__autores if a.codigo != autor.codigo]

    def incluirCapitulo(self, numeroCapitulo: int, tituloCapitulo: str):
        for cap in self.__capitulos:
            if cap.numeroCapitulo == numeroCapitulo:
                return
        novoCapitulo = Capitulo(numeroCapitulo, tituloCapitulo)
        self.__capitulos.append(novoCapitulo)

    def excluirCapitulo(self, tituloCapitulo: str):
        self.__capitulos = [
            c for c in self.__capitulos if c.tituloCapitulo != tituloCapitulo]

    def findCapituloByTitulo(self, tituloCapitulo: str):
        for cap in self.__capitulos:
            if cap.tituloCapitulo == tituloCapitulo:
                return cap
        return None


class Biblioteca:
    def __init__(self):
        self.__livros = []

    def incluirLivro(self, livro: Livro):
        if isinstance(livro, Livro) and livro not in self.__livros:
            self.__livros.append(livro)

    def excluirLivro(self, livro: Livro):
        self.__livros = [
            l for l in self.__livros if l.codigo != livro.codigo]

    @property
    def livros(self):
        return self.__livros
```

```python
import math


class Polinomio:
    def __init__(self, coeficientes):
        self.coeficientes = coeficientes
        self.grau = len(coeficientes) - 1

    def resultado(self, x):
        resultado = 0

        for N, coeficiente in enumerate(self.coeficientes):
            if N == 0:
                resultado += coeficiente
            else:
                resultado += coeficiente * math.pow(x, N)

        return resultado

    def soma(self, outro_polinomio):
        novos_coeficientes = []
        maior_grau = self.grau if self.grau > outro_polinomio.grau else outro_polinomio.grau

        for N in range(maior_grau + 1):
            soma_coeficiente = 0

            if N < self.grau + 1:
                soma_coeficiente += self.coeficientes[N]

            if N < outro_polinomio.grau + 1:
                soma_coeficiente += outro_polinomio.coeficientes[N]

            novos_coeficientes.append(soma_coeficiente)

        return Polinomio(novos_coeficientes)

    def subtrai(self, outro_polinomio):
        novos_coeficientes = []
        maior_grau = self.grau if self.grau > outro_polinomio.grau else outro_polinomio.grau

        for N in range(maior_grau + 1):
            coeficiente_a = 0
            coeficiente_b = 0

            if N < self.grau + 1:
                coeficiente_a = self.coeficientes[N]

            if N < outro_polinomio.grau + 1:
                coeficiente_b = outro_polinomio.coeficientes[N]

            novos_coeficientes.append(coeficiente_a - coeficiente_b)

        return Polinomio(novos_coeficientes)

    def multiplica(self, outro_polinomio):
        novos_coeficientes = [
            0]*(len(self.coeficientes) + len(outro_polinomio.coeficientes) - 1)

        for grauA, coefA in enumerate(self.coeficientes):
            for grauB, coefB in enumerate(outro_polinomio.coeficientes):
                novos_coeficientes[grauA + grauB] += coefA * coefB

        return Polinomio(novos_coeficientes)


def main():
    polinomio1 = Polinomio([-1, 1])
    polinomio2 = Polinomio([-6, 2, 1])
    soma_polinomios = polinomio1.soma(polinomio2)
    subtracao_polinomios = polinomio1.subtrai(polinomio2)
    multiplicacao_polinomios = polinomio1.multiplica(polinomio2)

    print(polinomio1.grau)
    print(polinomio1.resultado(8))
    print(soma_polinomios.coeficientes)
    print(subtracao_polinomios.coeficientes)
    print(multiplicacao_polinomios.coeficientes)


main()
```

### Exercício 4: Séries

Implemente uma classe que, implementa algumas séries matemáticas importantes:
Fibonacci, Fatorial, Fibonarial, Primo. Use recursão para Fibonacci e Fatorial.


```python
class Series:
    def __fibonacci__(self, numero):
        if numero > 1:
            return self.__fibonacci__(numero - 1) + self.__fibonacci__(numero - 2)

        return numero

    def fibonacci(self, limite):
        sequencia = []

        for i in range(limite):
            sequencia.append(self.__fibonacci__(i))

        return sequencia

    def __fatorial__(self, numero):
        if (numero == 1):
            return numero

        return numero * self.__fatorial__(numero - 1)

    def fatorial(self, numero):
        if numero < 0:
            return None
        elif numero == 0:
            return 1
        else:
            return self.__fatorial__(numero)

    def fibonarial(self, limite):
        sequencia = self.fibonacci(limite)
        nova_sequencia = []

        for numero in sequencia:
            nova_sequencia.append(self.fatorial(numero))

        return nova_sequencia

    def primos(self, final):
        sequencia = []

        for numero in range(2, final):
            primo = True

            for i in range(2, numero):
                if (numero % i == 0):
                    primo = False

            if primo:
                sequencia.append(numero)

        return sequencia

def main():
    serie = Series()
    print(serie.fibonacci(10))
    print(serie.fatorial(10))
    print(serie.fibonarial(10))
    print(serie.primos(10))

main()
```

### Exercício 10: Organização das Turmas
Crie um sistema que gerencia o cadastro de alunos e professores em turmas. Os usuários serão os membros da secretaria. Eles devem conseguir visualizar os alunos matriculados em cada turma, com seus dados, suas notas e presenças. Além disso, os secretários precisam ter acesso a dados cadastrais dos professores associados à disciplina.

```python
class Aluno:
    def __init__(self, nome, matricula, idade, serie, notas, lista_presenca):
        self.nome = nome
        self.matricula = matricula
        self.idade = idade
        self.serie = serie
        self.notas = notas
        self.lista_presenca = dict(lista_presenca)

    def media_por_disciplina(self, disciplina):
        soma = 0
        notas = self.notas[disciplina]
        quantidade = len(notas)
        for nota in notas:
            soma += nota
        return soma / quantidade

    def calcula_frequencia(self):
        soma = 0
        presencas = self.lista_presenca.values()
        total = len(presencas)

        for presenca in presencas:
            if (presenca):
                soma += 1

        return f"{100 * soma / total}%"


class Professor:
    def __init__(self, nome, disciplina):
        self.nome = nome
        self.disciplina = disciplina


class Turma:
    def __init__(self, serie, alunos, professores):
        self.serie = serie
        self.alunos = alunos
        self.professores = professores

    def resumo(self):
        print(f"Serie: {self.serie}ª \n")

        print(f"Alunos \n")
        for aluno in self.alunos:
            print(f"{aluno.nome} {aluno.matricula}")
            print(
                f"Media em História {aluno.media_por_disciplina('História')}")
            print(
                f"Media em Matemática {aluno.media_por_disciplina('Matemática')}")
            print(f"Frequência {aluno.calcula_frequencia()}")
            

        print(f"\nProfessores \n")
        for professor in self.professores:
            print(f"{professor.nome} {professor.disciplina}")


def main():
    fulano = Aluno("Fulano", 123456, 16, 2, {"História": [10, 8, 7, 9], "Matemática": [1, 2]},
                   {"12/04": True, "13/04": False, "14/04": False})
    ciclano = Aluno("Ciclano", 654321, 17, 2,  {"História": [10, 3, 1, 9], "Matemática": [10, 2]},
                    {"12/04": True, "13/04": True, "14/04": False})
    beltrano = Aluno("Beltrano", 654335, 15, 2,  {"História": [10, 9, 9, 9], "Matemática": [10, 10]},
                     {"12/04": True, "13/04": True, "14/04": True})

    alberto = Professor("Alberto", "História")
    rosana = Professor("Rosana", "Matemática")

    segundo_ano = Turma(2, [fulano, ciclano, beltrano], [alberto, rosana])

    segundo_ano.resumo()


main()
```
